```python
from sqlglot import parse_one
```

```python
from sqlgpt_parser.parser.mysql_parser import parser as mysql_parser
```

```python
from sqlparser import parse_statement
from sqlparser.utils import print_tree
```

### Read queries

```python
query = "SELECT first_name, last_name, salary FROM employees WHERE salary > 3800"
print(repr(parse_one(query)))
```

```python
print(repr(mysql_parser.parse(query)))
```

```python
print_tree(parse_statement(query))
```

```python
# join example
query = "SELECT table1.column1,table1.column2,table2.column1 FROM table1 LEFT JOIN table2 ON table1.matching_column = table2.matching_column;"
print(repr(parse_one(query)))
```

```python
mysql_parser.parse(query)
```

```python
# exists syntax use 
query = "SELECT emp_id FROM employees WHERE EXISTS (SELECT * FROM departments WHERE employees.department_id = departments.id);"
print(repr(parse_one(query)))
```

```python
# between syntax use
query = "SELECT * FROM employees WHERE salary BETWEEN 3000 AND 5000;"
print(repr(parse_one(query)))
```

```python
# like syntax use
query = "SELECT * FROM employees WHERE first_name LIKE 'A%';"
print(repr(parse_one(query)))
```

```python
# in syntax use
query = "SELECT * FROM employees WHERE department_id IN (1, 2, 3);"
print(repr(parse_one(query)))
```

```python
# wildcard with join syntax use
query = "SELECT * FROM employees e JOIN departments d ON e.department_id = d.id WHERE d.department_name LIKE 'Sales%';"
print(repr(parse_one(query)))
```

```python
# simple group by syntax sample 
query = "SELECT department_id, COUNT(*) FROM employees GROUP BY department_id;"
print(repr(parse_one(query)))
```

```python
# order by syntax sample 
query = "SELECT first_name, last_name FROM employees ORDER BY last_name ASC;"
print(repr(parse_one(query)))
```

```python
# advance query with "with", "group by", and "join" syntax
query = " WITH department_salaries AS (SELECT department_id, AVG(salary) AS avg_salary FROM employees GROUP BY department_id) SELECT d.department_name, ds.avg_salary FROM departments d JOIN department_salaries ds ON d.id = ds.department_id WHERE ds.avg_salary > 4000;"
print(repr(parse_one(query)))
```
