```python
from sqlglot import parse_one
```

```python
query = """SELECT date, amount FROM Transaction;"""
print(repr(parse_one(query)))
```

```python
query = """SELECT SUM(amount) AS total_order_amount FROM Order;"""
print(repr(parse_one(query)))
```

```python
query = "SELECT DATE_TRUNC('month', date) AS month, type, SUM(amount) AS total_amount FROM Transaction GROUP BY bank, type;"
print(repr(parse_one(query)))
```

```python
query = "SELECT AVG(amount) AS avg_loan_amount, AVG(duration) AS avg_duration FROM Loan;" 
print(repr(parse_one(query)))
```
